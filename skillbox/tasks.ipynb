# Задачі на дебіторську обертаємість

**Задача:**

Компанія "Дельта" має такі дані за звітний рік:
- Чисті продажі склали 3 800 000 грн.
- Дебіторська заборгованість на початок року становила 320 000 грн, а на кінець року — 370 000 грн.
- Крім цього, компанія надає знижку 2% клієнтам, які сплачують протягом 10 днів. В результаті, 30% клієнтів скористалися цією знижкою. 

**Задача 1:** Визначити коефіцієнт обертаємості дебіторської заборгованості, враховуючи, що 30% продажів були здійснені з наданням знижки.

**Задача 2:** Обчислити кількість днів для оберту дебіторської заборгованості.


```python
avgDebit = (320000+370000)/2
discount = 0.02
customersUsedDiscount = 0.3
customersDidntUseDiscount = 1-customersUsedDiscount
totalRevenue = 3800000
```

```python
# крок 1; Порахувати загальні продажі спираючись на дискаунт
totalSales = totalRevenue*customersUsedDiscount*(1-discount)+totalRevenue*customersDidntUseDiscount
```

```python
# крок 2; Порахувати обертаємість
totalSales/avgDebit
```

```python
# Обертаємість у днях
365/(totalSales/avgDebit)
```
