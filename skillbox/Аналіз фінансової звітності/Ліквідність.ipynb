### Основні аспекти ліквідності:

**Ліквідність активів:** Це стосується того, як швидко та легко можна продати активи (наприклад, готівка, цінні папери, запаси) без значної втрати їхньої вартості. Найбільш ліквідними активами є готівка та банківські депозити.

**Ліквідність підприємства:** Це здатність компанії виконувати свої короткострокові зобов'язання. Для оцінки ліквідності підприємства використовуються різні фінансові показники, зокрема:овими зобов'язаннями.овими зобов'язаннями.

$$ \text {Коефіцієнт поточної ліквідності} = \frac{{\text {Оборотні активи}}}{{\text {поточні забовязання}}} $$
$$ \text {Коефіцієнт поточної ліквідності} = \frac{{\text {Оборотні активи - Запаси}}}{{\text {поточні забовязання}}} $$

Ось прості задачі на ліквідність та швидку ліквідність:

### 1. Задача на ліквідність
**Умови:**  
Компанія "К" має оборотні активи у розмірі \\$300,000 та поточні зобов'язання у розмірі \\$150,000.  
**Запитання:** Який коефіцієнт поточної ліквідності у компанії?

**Рішення:**  
Коефіцієнт поточної ліквідності = Оборотні активи / Поточні зобов'язання  
Коефіцієнт поточної ліквідності = $300,000 / $150,000 = 2.0
