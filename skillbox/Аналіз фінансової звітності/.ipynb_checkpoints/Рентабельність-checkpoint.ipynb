# Основні види рентабельності:
1. **Рентабельність активів (ROA, Return on Assets)** — показує, скільки прибутку генерується на кожну одиницю активів підприємства:
   $$ ROA = \frac{\text{Чистий прибуток}}{\text{Середні активи}} \times 100\%\ $$
2. **Рентабельність власного капіталу (ROE, Return on Equity)** — відображає прибутковість вкладеного власниками капіталу:
   $$ ROE = \frac{\text{Чистий прибуток}}{\text{Власний капітал}} \times 100\% $$
3. **Рентабельність продажів (ROS, Return on Sales)** — показує частку прибутку в кожній одиниці доходу від продажів:
   $$ ROS = \frac{\text{Чистий прибуток}}{\text{Дохід від продажів}} \times 100\% $$


### 1. Рентабельність активів (ROA)
Компанія отримала чистий прибуток у розмірі \\$200,000. <br> 
Середні активи компанії за звітний період становлять \\$1,000,000. <br>
Обчисли рентабельність активів.

```python
profit = 200_000
actives = 1_000_000
```

```python
ROA = profit/actives*100
```

```python
ROA
```

### 2. Рентабельність власного капіталу (ROE)
Компанія має чистий прибуток у розмірі \\$150,000, <br>
а власний капітал становить \\$500,000. <br>
Обчисли рентабельність власного капіталу.

```python
profit2 = 150_000
own_capital = 500_000
```

```python
ROE = profit2/own_capital*100
```

```python
ROE
```

### 3. Рентабельність продажів (ROS)
Компанія заробила чистий прибуток \\$50,000 <br>
при доході від продажів у розмірі \\$500,000. <br>
Обчисли рентабельність продажів.

```python
profit3 = 50_000
revenue = 500_000
```

```python
ROS = profit3/revenue*100
```

```python
ROS
```

### 1. Рентабельність активів (ROA)
#### Задача:
Компанія "Г" отримала чистий прибуток у розмірі \\$250,000. <br> Середні активи за рік становлять \\$1,500,000, <br> але компанія також здійснила капітальні інвестиції в розмірі \\$300,000 на нове обладнання. <br> Обчисли рентабельність активів, враховуючи нові інвестиції.

```python
profit = 250_000
investition = 300_000
actives = 1_500_000

all_actives = investition+actives
ROA = round(profit/all_actives*100, 2)
ROA
```

### 2. Рентабельність власного капіталу (ROE)
#### Задача:
Компанія "Д" має чистий прибуток у розмірі \\$120,000. <br> Загальні зобов'язання компанії становлять \\$600,000, а активи — \\$1,000,000. <br> Обчисли власний капітал компанії та рентабельність власного капіталу.

```python
profit = 120_000
actives = 1_000_000
debts = 600_000

ROE = profit/(actives-debts)*100
ROE
```

### 3. Рентабельність продажів (ROS)
#### Задача:
Компанія "Е" отримала чистий прибуток у розмірі \\$80,000 <br> при загальному доході від продажів \\$600,000. <br> Однак, для досягнення цього доходу компанія витратила \\$20,000 на рекламу <br> та \\$50,000 на операційні витрати.<br> Обчисли рентабельність продажів з урахуванням витрат на рекламу та операційних витрат.

```python
profit = 80_000
revenue = 600_000
spends = 20_000+50_000

ROS = round((profit-spends)/revenue*100, 2)
ROS
```
