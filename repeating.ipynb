```python
avg_debit = (debit_start+debit_finish)/2
```

```python
365/(real_sells/avg_debit)
```

Компанія надає знижку 5% клієнтам, які сплачують протягом 10 днів, і 40% клієнтів скористалися цією знижкою.
Також компанія має частину клієнтів, які затримують оплату більше 60 днів, і їх частка складає 10% від загального обсягу продажів.
Задача 1: Визначити коефіцієнт обертаємості дебіторської заборгованості, враховуючи, що 40% продажів були здійснені зі знижкою, а 10% клієнтів затримали оплату.

Задача 2: Обчислити кількість днів для одного оберту дебіторської заборгованості.

```python
net_sales = 8_200_000
debit_start = 600_000
debit_finish = 720_000
discount = 0.05
customers_used_discount = 0.4
delay = 0.1
```

```python
avg_debit = (debit_start+debit_finish)/2
total_sales = (net_sales*(1-discount)*customers_used_discount)+(net_sales*(1-customers_used_discount))-net_sales*delay
avg_discount = total_sales/avg_debit
avg_discount_days = 365/(total_sales/avg_debit)
avg_discount, avg_discount_days
```
